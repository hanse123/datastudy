# 지역아동센터 

```python
import numpy as np
import pandas as pd
```

```python
cc = pd.read_csv('공공빅데이터 프로젝트 - 장난감도서관/서울시 지역아동센터 시설현황정보.csv',
                                thousands=',',
                                encoding='euc-kr')
```

```python
cc.head()
```

```python
cc = cc.dropna(subset=["지번주소"])
```

```python
cc.isnull().sum()
```

```python
cc.info()
```

```python
cc['지번주소']
```

```python
dong = []
for i in cc['지번주소'] :
    a = i.split()
    dong.append(a[2])
dong
```

```python
from pandas import DataFrame
```

```python
ccenter = DataFrame()
```

```python
ccenter['시설명'] = cc['시설명']
ccenter['지번주소'] = cc['지번주소']
ccenter['lat'] = cc['X좌표값']
ccenter['lng'] = cc['Y좌표값']
ccenter['구'] =  cc['자치구명']
ccenter['동'] = dong
```

```python
ccenter
ccenter.to_csv('공공빅데이터 프로젝트 - 장난감도서관/ccenter.csv' ,sep=',', na_rep='NaN')
```

```python
freq_ccenter = ccenter.groupby(['동']).count() 
freq_ccenter.drop(['지번주소','lat','lng','구'], axis = 1, inplace = True)
freq_ccenter.head()
freq_ccenter.to_csv('공공빅데이터 프로젝트 - 장난감도서관/freq_ccenter.csv' ,sep=',', na_rep='NaN')
```

# 유치원

```python
import pandas as pd
```

```python
kin = pd.read_csv('공공빅데이터 프로젝트 - 장난감도서관/서울특별시 유치원 일반현황.csv',
                                thousands=',',
                                encoding='euc-kr')
kin.head()
```

```python
kin = kin.dropna(subset=["지번주소"])
```

```python
kin.isnull().sum()
```

```python
kin.info()
```

```python
kin['지번주소']
```

```python
gu = []
for i in kin['지번주소'] :
    a = i.split()
    gu.append(a[1])
gu
```

```python
from pandas import DataFrame
```

```python
kindergarten = DataFrame()
```

```python
kindergarten['유치원명'] = kin['유치원명']
kindergarten['지번주소'] = kin['지번주소']
kindergarten['주소'] = kin['주소']
kindergarten['구'] = gu
```

```python
kindergarten
```

```python
kindergarten.to_csv('공공빅데이터 프로젝트 - 장난감도서관/kindergarten.csv' ,sep=',', na_rep='NaN')
```

```python
freq_kindergarten = kindergarten.groupby(['구']).count() 
freq_kindergarten.drop(['지번주소','주소'], axis = 1, inplace = True)
freq_kindergarten.to_csv('공공빅데이터 프로젝트 - 장난감도서관/freq_kindergarten.csv' ,sep=',', na_rep='NaN', encoding = 'euc-kr')
```
